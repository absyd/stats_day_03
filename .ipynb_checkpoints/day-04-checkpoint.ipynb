```python
import numpy as np
from scipy.stats import ttest_ind


# ABTest Framework
class ABTest:
    def __init__(self, A, B, alpha=0.05):
        self.A = np.array(A)
        self.B = np.array(B)
        self.alpha = alpha

    def summary(self):
        return {
            "mean_A": np.mean(self.A),
            "mean_B": np.mean(self.B),
            "std_A": np.std(self.A),
            "std_B": np.std(self.B),
            "n_A": len(self.A),
            "n_B": len(self.B)
        }

    def run_test(self):
        stat, p = ttest_ind(self.A, self.B)
        return stat, p

    def decision(self):
        stat, p = self.run_test()
        if p < self.alpha:
            return f"Reject Null Hypothesis H0 (p={p:.4f}) → Significant difference"
        else:
            return f"Fail to reject H0 (p={p:.4f}) → No strong evidence"

    def report(self):
        print("Summary:", self.summary())
        stat, p = self.run_test()
        print(f"T-stat: {stat:.4f}, P-value: {p:.4f}")
        print(self.decision())
```

```python
A = np.random.binomial(1, 0.10, 1000)  # 10% conversion
B = np.random.binomial(1, 0.12, 1000)  # 12% conversion

test = ABTest(A, B)
test.report()
```
